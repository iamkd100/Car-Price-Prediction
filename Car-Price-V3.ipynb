```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Car details v3.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
# Encode Fuel Data
df.replace({'fuel':{'Diesel': 0, 'Petrol': 1, 'CNG': 2, 'LPG': 3}}, inplace=True)

# Encode Transmission
df.replace({'transmission':{'Manual': 0, 'Automatic': 1}}, inplace=True)

# Encode Seller Type Data
df.replace({'seller_type':{'Individual': 0, 'Dealer': 1, 'Trustmark Dealer': 2}}, inplace=True)

# Encode Owner Data
df.replace({'owner':{'First Owner': 0, 'Second Owner': 1, 'Third Owner': 2, 'Fourth & Above Owner': 3, 'Test Drive Car': 4}}, inplace=True)
```

```python
df.loc[:, ['fuel', 'transmission', 'seller_type', 'owner']]
```

```python
df.info()
```

```python
df.loc[:, ['mileage', 'engine', 'max_power', 'torque']]
```

```python
df['mileage']=df['mileage'].str.split(' ').str.get(0).str.replace(',','')
df['engine']=df['engine'].str.split(' ').str.get(0).str.replace(',','')
df['max_power']=df['max_power'].str.split(' ').str.get(0).str.replace(',','')
df
```

```python
df['mileage']=df['mileage'].astype(float)
df['engine']=df['engine'].astype(int)
df['max_power'] = pd.to_numeric(df['max_power'])
df.info()
```

```python
corr = df.corr()
corr
```

```python
plt.figure(dpi=110)
sns.heatmap(corr, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```

```python
df.dtypes
```

```python
X = df.drop(columns=['selling_price','name','torque'],axis=1)
Y = df['selling_price']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = df['selling_price']
X1 = scaler.fit_transform(X)
X = pd.DataFrame(data= X1, columns = X.columns)
X.head()
```

```python
X.corr()
```

```python
k = X.corr()
z = [[str(i),str(j)] for i in k.columns for j in k.columns if (k.loc[i,j] >abs(0.5))&(i!=j)]
z, len(z)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = X[:]

VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(x_train, y_train)
```

```python
predictions = RF.predict(x_test)
RF.score(x_test, y_test)
```

```python
df = pd.DataFrame({'Actual Values':y_test, 'Predicted Values':predictions})
df
```

```python
test = pd.DataFrame({'Predicted value':predictions, 'Actual value':y_test})
fig= plt.figure(figsize=(16,8))
test = test.reset_index()
test = test.drop(['index'],axis=1)
plt.plot(test[:50])
plt.legend(['Actual value','Predicted value'])
```
